## Comparison of various sampling methods on PCA2 plot

```python
n = 200
def sample_keeping_pop_breadth(labels, n):
    '''
    Take in the popinfo.csv file and return a list of indices representing a sample of points
    which maintains the breadth in population. It does this by taking a single row, without replacement
    from each population until there are n indices.
    '''
    result = []

    df = labels.copy()

    while True:
        if df['POP'].value_counts().count() < n:
            n -= df['POP'].value_counts().count()
            # Add one of each
            for pop in df['POP'].unique():
                idx = df[df['POP']==pop].sample(1).index
                result.extend(idx)
                df = df.drop(idx)
        else:
            # sample from remainig ones
            result.extend(df.sample(n).index)
            break
            
    return result
```

```python
n = 400

weights = (1 / labels['POP'].value_counts()).rename("weights")

sampled = labels.join(pd.DataFrame(data = weights),on='POP').sample(n=n, weights='weights')
sampled_indices = sampled.index
sampled_indices
```

```python
pca2.loc[:n,['PC1','PC2']].plot.scatter('PC1','PC2')
plt.title('PCA 2D for Himalayan Data, sampled by order')

pca2.loc[sampled_indices,['PC1','PC2']].plot.scatter('PC1','PC2')
plt.title('PCA 2D for Himalayan Data, sampled by inverse frequency')

pca2.loc[sample_keeping_pop_breadth(labels,n),['PC1','PC2']].plot.scatter('PC1','PC2')
plt.title('PCA 2D for Himalayan Data, sampled to keep population breadth')

pca2.loc[:,['PC1','PC2']].plot.scatter('PC1','PC2')
plt.title('PCA 2D for Himalayan Data')

plt.show()
```
