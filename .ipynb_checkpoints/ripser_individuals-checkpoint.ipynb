```python
gt_matrix_PCs_2 = np.load('/home/harryem/data/HimalGen/gt_matrix_PCs_2.npy').T.astype(float)
```

```python
print(gt_matrix_PCs_2.shape)
gt_matrix_PCs_2_ave = gt_matrix_PCs_2.reshape(-1, 2, 2).mean(axis=1)
print(gt_matrix_PCs_2_ave.shape)
```

```python
# Get individuals in top K cocycles

K = 20

# individuals_list = []
individuals_df = pd.DataFrame(columns=['cocycle_idx'])

for i, idx in enumerate(ordered_cocyles[:K]):
    cocycle = cocycles[1][idx]
    individuals = set(cocycle[:,:2].flatten() // 2)
#     individuals_list.append(labels.loc[individuals,'IID'].values)
    for indiv in individuals:
        if indiv in individuals_df.index:
            # Some individuals are in more than one cocyle. The most persistent is chosen
            print("Dupe {}".format(indiv))
        else:
#             individuals_df.loc[indiv] = idx
            individuals_df.loc[indiv] = i
```

```python
individuals_df = individuals_df#.astype(int)
```

```python
pcs_cocycles = pd.DataFrame(gt_matrix_PCs_2_ave, columns = ['PC1', 'PC2']).join(individuals_df, how='left')
pcs_cocycles.loc[pcs_cocycles.isnull().cocycle_idx, 'cocycle_idx'] = 10000
# pcs_cocycles['cocycle_idx'].unique()
# pcs_cocycles.isnull().cocycle_idx
```

```python
pcs_cocycles.plot.scatter('PC1', 'PC2', c = 'cocycle_idx')
```
